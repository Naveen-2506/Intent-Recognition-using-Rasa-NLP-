```python
# Load the Packages
from rasa_nlu.training_data  import load_data
from rasa_nlu.config import RasaNLUModelConfig
from rasa_nlu.model import Trainer
from rasa_nlu import config
```

```python
#load the data
!cat YPO_New_Trained_Intents.json
```

```python
# Loading the DataSet
train_data = load_data('YPO_New_Trained_Intents.json')
```

```python
# Config a Backend using Sklearn and Spacy
trainer = Trainer(config.load("config_spacy.yaml"))
```

```python
# Training Data
trainer.train(train_data)
```

```python
# Returns the directory the model is stored in (Creat a folder to store model in)
model_directory = trainer.persist('/home/naveen/Documents/ypo project/intent recognition/realprojects/')
```

```python
#Making Predictions With Model
from rasa_nlu.model import Metadata, Interpreter
```

```python
# where `model_directory points to the folder the model is persisted in
interpreter = Interpreter.load(model_directory)
```

```python
# Using strip and split function remove whitespace characters issue.
# And interpreter.parse Declare in search
def search(Searchsent):
    Searchsent=Searchsent.strip()
    Searchsent=" ".join(Searchsent.split())
    return(interpreter.parse(Searchsent))
```
